```python
class TrieNode:
    def __init__(self):
        self.children = {}
        self.is_end = False


class Trie:
    def __init__(self):
        self.root = TrieNode()

    # Insert contact
    def insert(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                node.children[char] = TrieNode()
            node = node.children[char]
        node.is_end = True

    # Search complete word
    def search(self, word):
        node = self.root
        for char in word:
            if char not in node.children:
                return False
            node = node.children[char]
        return node.is_end

    # Check prefix
    def starts_with(self, prefix):
        node = self.root
        for char in prefix:
            if char not in node.children:
                return False
            node = node.children[char]
        return True


# -------- Main Program --------

n = int(input("Enter number of contacts: "))
trie = Trie()

for _ in range(n):
    name = input("Enter contact name: ")
    trie.insert(name)

print('Contact "Anil" found →', trie.search("Anil"))
print('Contact "Anand" not found →', trie.search("Anand"))
print('Prefix "Ani" exists →', trie.starts_with("Ani"))
print('Prefix "Bal" exists →', trie.starts_with("Bal"))
```
