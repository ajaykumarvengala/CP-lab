```python
def max_subarray(arr, left, right):
    # Base case: one element
    if left == right:
        return arr[left]

    mid = (left + right) // 2

    # Max in left half
    left_max = max_subarray(arr, left, mid)

    # Max in right half
    right_max = max_subarray(arr, mid + 1, right)

    # Max crossing sum
    cross_max = max_crossing_sum(arr, left, mid, right)

    # Return maximum of three
    return max(left_max, right_max, cross_max)

def max_crossing_sum(arr, left, mid, right):
    current_sum = 0
    left_sum = -float('inf')  # Equivalent to Long.MIN_VALUE

    # From mid to left
    for i in range(mid, left - 1, -1):
        current_sum += arr[i]
        left_sum = max(left_sum, current_sum)

    current_sum = 0
    right_sum = -float('inf')  # Equivalent to Long.MIN_VALUE

    # From mid+1 to right
    for i in range(mid + 1, right + 1):
        current_sum += arr[i]
        right_sum = max(right_sum, current_sum)

    return left_sum + right_sum

# Main execution block
if __name__ == '__main__':
    T = int(input()) # Read the number of test cases

    results = []
    for _ in range(T):
        N = int(input()) # Read the size of the array
        # Read array elements as space-separated integers on a single line
        arr = list(map(int, input().split()))

        result = max_subarray(arr, 0, N - 1)
        results.append(result)

    for res in results:
        print(res)
```
